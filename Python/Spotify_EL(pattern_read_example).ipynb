```python
#Import library
import pandas as pd
import os
import numpy as np
import sqlalchemy
import json
from glob import glob
```

```python
json_dir = 'C:/Users/Jason/OneDrive/Desktop/belajar/spotify project/MyData/'

json_pattern = os.path.join(json_dir, '*.json')
file_list = glob(json_pattern)
```

```python
file_list
```

```python
dfs = []
for file in file_list:
    with open(file, encoding='utf8') as f:
        json_data = pd.json_normalize(json.loads(f.read()))
    dfs.append(json_data)
df = pd.concat(dfs)
```

```python
df.head(1)
```

```python
engine = sqlalchemy.create_engine("postgresql+psycopg2://jason:jason@localhost:2700/spotify_project")
#connect database to postgresql server 
```

```python
df.to_sql('spotify_project',engine,if_exists='append')
```
